```python
data.rename(columns={"annual_income":"Annual Income"},inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
cat = []
for c in  data.columns:
  if data[c].dtype=="object":
    cat.append(c)
cat
```

```python
for c in cat:
  data[c]=lb.fit_transform(data[c])
data
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
X = data.drop("Annual Income",axis=1)
X.columns
```

```python
Y = data["Annual Income"]
Y
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
clf = LogisticRegression(max_iter=300)
clf.fit(train_x,train_y)
```

```python
perd_y=clf.predict(test_x)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
confusion_matrix(test_y,perd_y)
```

```python
accuracy_score(test_y,perd_y)
```

```python
classification_report(test_y,perd_y)
```
